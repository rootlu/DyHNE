```python
import numpy as np
import pandas as pd
import os
```

```python
data_dir = '../data/dblp'
pa_data = pd.read_csv(os.path.join(data_dir,'paper_author.txt'),header=None,sep='\t')
pa_data
```

```python
pc_data = pd.read_csv(os.path.join(data_dir,'paper_conf.txt'),header=None,sep='\t')
pc_data
```

```python
len(set(pa_data[0])), len(set(pa_data[1]))
```

```python
A_pa_data = pa_data.copy()
A_pa_data
```

```python
A_pa_data.iloc[:, 1] = A_pa_data.iloc[:, 1].apply(lambda x:x+14376)
```

```python
np.array(A_pa_data[1]).tolist()+np.array(A_pa_data[0]).tolist()
```

```python
from scipy.sparse import csr_matrix
n = len(set(A_pa_data[0]))+len(set(A_pa_data[1]))
row = np.array(A_pa_data[0]).tolist()+np.array(A_pa_data[1]).tolist()
col = np.array(A_pa_data[1]).tolist()+np.array(A_pa_data[0]).tolist()
data = np.ones(len(row))
pa_mtx = csr_matrix((data, (row, col)), shape=(n, n))
```

```python
pa_mtx.shape
```

```python
import scipy.io
scipy.io.savemat(os.path.join(data_dir,'pa'),{"pa":pa_mtx})
```

```python
A_pc_data = pc_data.copy()
n_p = len(pc_data[0])
A_pc_data.iloc[:, 1] = A_pc_data.iloc[:, 1].apply(lambda x:x+n_p)
n_pc = len(set(A_pc_data[0]))+len(set(A_pc_data[1]))
row = np.array(A_pc_data[0]).tolist()+np.array(A_pc_data[1]).tolist()
col = np.array(A_pc_data[1]).tolist()+np.array(A_pc_data[0]).tolist()
data = np.ones(len(row))
pc_mtx = csr_matrix((data, (row, col)), shape=(n_pc, n_pc))
A_pc_data
```

```python
scipy.io.savemat(os.path.join(data_dir,'pc'),{"pc":pc_mtx})
```
